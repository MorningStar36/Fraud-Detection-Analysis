```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_rows',4)
print("Setup Complete")
```

```python
iris_data = pd.read_csv('./data/iris.csv', index_col='Id')
iris_data.head()
```

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
sns.kdeplot(iris_data['Petal Length (cm)'], fill=True)
```

```python
plt.figure(figsize=(20,40))
sns.jointplot(x=iris_data['Petal Length (cm)'] , y=iris_data['Sepal Length (cm)'], kind="kde")
```

```python
sns.histplot(data=iris_data, x='Petal Length (cm)' , hue='Species')
```

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue = 'Species')
```
